```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def implicit_euler_solve(u0, T, dt):
  num_steps = int(T/dt)
  tt = np.arange(num_steps+1)*dt
  u = np.empty(num_steps+1)
  u[0] = u0
  for k in range(num_steps):
      u[k+1] = (u[k] + np.sqrt(u[k]**2 + 4*dt))/2
  return tt, u
```

```python
T = 10
dt = 0.1
u0 = 1.0

plt.figure(figsize=(18, 9))
tt, u = implicit_euler_solve(u0, T, dt)
plt.plot(tt, u, 'o--', label='numeric solution')

plt.plot(tt, np.sqrt(2*tt + 1), '-', lw=2, label='ground truth')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
